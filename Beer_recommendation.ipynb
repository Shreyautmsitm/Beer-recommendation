```python
# Setting web driver
wd = webdriver.Chrome('chromedriver',options=chrome_options)
```

```python
import pandas as pd

# # Variables 
count = 0
comments = pd.DataFrame(columns = ['product_name','product_review','user_rating'])

# # Getting Initial Review Pages 
routes = [] # holds all the review pages 
names = [] # beer names 
route = 'https://www.beeradvocate.com/beer/top-rated/'
wd.get(route)

# # Getting Rows of the 250 Beers Table 
rows = 1 + len(wd.find_elements_by_xpath("//*[@id='ba-content']/table/tbody/tr/td/a"))

print("Getting Beer names")
for r in range(2,220): # Getting 220 Beers 
  element = "//*[@id='ba-content']/table/tbody/tr[{}]/td[2]/a".format(r)
  value = wd.find_element_by_xpath(element).get_attribute("href") 
  element2 = "//*[@id='ba-content']/table/tbody/tr[{}]/td[2]/a/b".format(r)
  names.append(wd.find_element_by_xpath(element2).text)
  
  routes.append(value) # Adding values to the list of links to individual reviews page

print("Loaded Beer Names: ", len(routes))
# Going thru individual reviews page
for idx,route in enumerate(routes): 

  # Fetching routes 
  wd.get(route)

  # Getting ids  
  ids = wd.find_elements_by_xpath("//*[@id='rating_fullview_container']")
   
  # Looping through the page
  print(names[idx],"  ",len(ids))
  for x in range(0, len(ids)):
      # Product Name 
      product_name = names[idx]
      
      # Product Rating 
      user_rating = wd.find_elements_by_xpath('//*[@id="rating_fullview_content_2"]/span[2]')[x].text
      #print(product_rating)

      # Product Review
      product_review = wd.find_elements_by_xpath('//*[@id="rating_fullview_content_2"]')[x].text
      for i in range(1,7): 
        span_elem = '//*[@id="rating_fullview_content_2"]/span[{}]'.format(i)
        if wd.find_elements_by_xpath(span_elem)[x].text is not None:
          span_for_removal = wd.find_elements_by_xpath(span_elem)[x].text
          product_review = product_review.replace(span_for_removal,"")
      product_review = product_review.replace('rDev', "")
      product_review = product_review.replace('\n', " ")
      product_review = product_review.replace('/', "")
      product_review = product_review.replace('look:  | smell:  | taste:  | feel:  | overall: ',"")    
      # Adding data
      comments.loc[len(comments)] = [product_name,product_review,user_rating]
      count +=1 
  # Moving onto the next page
#  print(count)

# Final output
#print(comments)
#print(comments.shape[0])
print('Final: ' + str(count))
```

```python
#Cleaning the review file and removing the stop words
reviews_df = reviews_df.dropna()
reviews_df['cleaned_review'] = reviews_df['product_review'].apply(lambda x :x.translate(str.maketrans('', '', string.punctuation))) # remove punctuation
reviews_df['cleaned_review'] = reviews_df['cleaned_review'].apply(lambda x :x.lower())                                              # make lower
reviews_df['cleaned_review'] = reviews_df['cleaned_review'].apply(word_tokenize).apply(set).apply(list)                    # make a unique list of words

def remove_stopwords(s):
    return [w for w in s if not w in stop_words] 
    
reviews_df['cleaned_review'] =  reviews_df['cleaned_review'].apply(remove_stopwords)
```

```python
#Reading the attributes.csv file
attributes_csv = pd.read_csv("attributes.csv" )
print(attributes_csv)
attributes_csv= attributes_csv['Attributes'].tolist()
print(attributes_csv)
```

```python
# choosing only product name and reviews from the original dataframe
col_names = ['product_name','product_review' ]
Taskc_op= reviews_df[col_names]
print(Taskc_op)
```

```python
# cosine similarity function
import numpy as np

def cosine_similarity_calc(vec_1,vec_2):
	cosine = np.dot(vec_1,vec_2)/(np.linalg.norm(vec_1)*np.linalg.norm(vec_2))
	return cosine
```

```python
# make an array of tokenized words to vectorize for cosine similarity calculation
reviews_token= reviews_df['cleaned_review']
reviews_token = np.array(reviews_token)
print(reviews_token[0])
```

```python
# vectorize tokens and attributes and calculate cosine similarity
cosine_final =[]
for i in range(len(reviews_token)):
  vec_1 = []
  vec_2 = []
  
  #print(reviews_token[0])
  X_set = {w for w in reviews_token[i] if not w in reviews_token}
  Y_set = {w for w in attributes_csv }
  #print("X_Set is " ,X_set)
  #print("Y_Set is " ,Y_set)
  vector = X_set.union(Y_set) 
  for w in vector:
    if w in X_set: 
      vec_1.append(1) 
    else: 
      vec_1.append(0)
    if w in Y_set: 
      vec_2.append(1)
    else: 
      vec_2.append(0)
  cosine = cosine_similarity_calc(vec_1, vec_2)
  #print(cosine)
  cosine_final.append(cosine)
#print(cosine_final)
col3 = pd.DataFrame(cosine_final)
#print(col3)
```

```python
# Writing the code in to a csv file

Taskc_op['similarity_score'] = col3
print(Taskc_op)
Taskc_op.to_csv("Taskc_op.csv", index = None)
!ls
from google.colab import files
files.download("Taskc_op.csv")
```

# **CONCLUSION**
###**Why are the recommendations different?**

The recommendations changes quite a bit among the task -E, F & G , this is mainly due to the different parameters used to get the recommendations.Products recommended in task G do not fit well with our chosen attributes because they were recommended solely based on average user rating. Average user rating does not guarantee that the user will favor the product. When a product is recommended to the user, it is important to consider similarity and sentiment scores because it gives more accurate recommendations based on products the user already likes. Sentiment scores take care of the subjectivity involved in an individual's decision making process. Similarity and sentiment scores together prove how similar the product is to the one that the user picked out and the favorability of the product reviews. Also, as observed from Task F, the top-3 beers based on word vector approach is much different than the bag of words. The sentiment scores and similarity numbers are not always correlated. As shown in "recommendations based on spacy similarity", recommending beers purely based off of similarity scores and disregarding sentiment,Saison Bernice will be in the top-3. The sentiment score for Saison Bernice clearly shows a negative preference. Therefore, recommendations for beers in either bag of words or word vector approaches should be accompanied by sentiment evaluation.
